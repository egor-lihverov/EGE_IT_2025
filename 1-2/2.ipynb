```python
def F(w, x, y, z):
    return (w != y) and (not z or w) and not x


print("w\t x\t y\t z\t F")


for w in [0, 1]:
    for x in [0, 1]:
        for y in [0, 1]:
            for z in [0, 1]:
                if F(w, x, y, z):
                    print(f"{w}\t {x}\t {y}\t {z}\t {int(F(w, x, y, z))}")
```
