Область наблюдения представляет собой круг с центром в точке (0, 0). <p>Истинный центр кластера, или <strong>центроид</strong>, — это одна из звёзд на графике, сумма расстояний от которой до всех остальных звёзд кластера минимальна.</p>

<p><strong>Справка о полярной системе координат:</strong></p>
<p>Полярная система координат — это двумерная система координат, в которой каждая точка на плоскости определяется двумя числами: полярным углом и полярным радиусом.</p>
<ul>
<li>Полярная ось — это фиксированная прямая, от которой отсчитывается полярный угол. Важно отметить, что полярная ось совпадает с положительным направлением оси абсцисс (оси <span class="math math-inline"><span class="katex"><span class="katex-mathml"><math xmlns="http://www.w3.org/1998/Math/MathML"><semantics><mrow><mi>X</mi></mrow><annotation encoding="application/x-tex">X</annotation></semantics></math></span><span class="katex-html" aria-hidden="true"><span class="base"><span class="strut" style="height: 0.6833em;"></span><span class="mord mathnormal" style="margin-right: 0.07847em;">X</span></span></span></span></span>) в декартовой системе координат.</li>
<li>Полярный угол (<span class="math math-inline"><span class="katex"><span class="katex-mathml"><math xmlns="http://www.w3.org/1998/Math/MathML"><semantics><mrow><mi>ϕ</mi></mrow><annotation encoding="application/x-tex">\phi</annotation></semantics></math></span><span class="katex-html" aria-hidden="true"><span class="base"><span class="strut" style="height: 0.8889em; vertical-align: -0.1944em;"></span><span class="mord mathnormal">ϕ</span></span></span></span></span>) — это угол между полярной осью и линией, соединяющей точку с началом координат. Угол отсчитывается против часовой стрелки от полярной оси.</li>
<li>Полярный радиус (<span class="math math-inline"><span class="katex"><span class="katex-mathml"><math xmlns="http://www.w3.org/1998/Math/MathML"><semantics><mrow><mi>r</mi></mrow><annotation encoding="application/x-tex">r</annotation></semantics></math></span><span class="katex-html" aria-hidden="true"><span class="base"><span class="strut" style="height: 0.4306em;"></span><span class="mord mathnormal" style="margin-right: 0.02778em;">r</span></span></span></span></span>) — это расстояние от точки до начала координат.</li>

```python
import math

a = []

with open("./4/27_A.txt") as f:
    f.readline()
    for i in f.readlines():
        a.append(list(map(float, i.replace(',', '.').split())))

clust = [[], []]

for i in a:
    # print(i)
    x, y, h = i
    clust_num = 0 if x < 150 and y < 250 else 1 if x > 350 and y > 250 else 2
    if clust_num == 2: continue
    clust[clust_num].append(i)


def dist(a, b):
    return math.sqrt((a[0] - b[0]) ** 2 + (a[1] - b[1]) ** 2) * b[2]

def sum_dist(a, clust):
    return sum(dist(a, j) for j in clust)

ans = []

for j in range(len(clust)):
    mini = 1e18
    v = -1
    for i in range(len(clust[j])):
        val = sum_dist(clust[j][i], clust[j])
        if val < mini:
            mini = val
            v = clust[j][i]
    ans.append(v)
    
print(sum(i[0] for i in ans) / 2 * 100000, sum(i[1] for i in ans) / 2 * 100000)
```

```python
import math

a = []

with open("./5/27_B (1).txt") as f:
    f.readline()
    for i in f.readlines():
        a.append(list(map(float, i.replace(',', '.').split())))

clust = [[], [], []]

for i in a:
    clust_num = 2 if i[0] > 5 else 1 if i[1] > 7.2 else 0
    clust[clust_num].append(i)

def dist(a, b):
    return math.sqrt((a[0] - b[0]) ** 2 + (a[1] - b[1]) ** 2)

def sum_dist(a, clust):
    return sum(dist(a, j) for j in clust)

ans = []

for j in range(len(clust)):
    mini = 1e9
    v = -1
    for i in range(len(clust[j])):
        val = sum_dist(clust[j][i], clust[j])
        if val < mini:
            mini = val
            v = clust[j][i]
    ans.append(v)
    
print(sum(i[0] for i in ans) / 3, sum(i[1] for i in ans) / 3)
```
