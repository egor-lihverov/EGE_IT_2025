```python
def delit(n):
    d = []
    k = 2 #  1  ,    1  n
    while k**2 <= n:
        if n%k == 0:
            d.append(k)
            if n//k != k:
                d.append(n//k)
        k += 1
    return d
def m(n):
    d=delit(n)
    if len(d) < 2:
        return 0
    else:
        d.sort()
        return d[-1] + d[-2]
k = 0
n = 10000001
while k<5:
    mn = m(n)
    if 0 < mn < 10000:
        k += 1
        print(n, mn)
        print(delit(n))
    n += 1
```
