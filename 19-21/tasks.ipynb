## 19

<div class="Markdown-module__Markdown--f5ALc TasksBankTaskCard-module__TaskCardContentMarkdown--hsXqP"><p>Два игрока, Петя и&nbsp;Ваня, играют в&nbsp;следующую игру. Перед игроками лежит куча камней. Игроки ходят по&nbsp;очереди, первый ход делает Петя. За&nbsp;один ход игрок может добавить в&nbsp;кучу 3&nbsp;камня, 10&nbsp;камней или увеличить количество камней в&nbsp;2&nbsp;раза. У&nbsp;каждого игрока есть неограниченное количество камней, чтобы делать ходы.</p>
<p>Игра завершается в&nbsp;тот момент, когда количество камней в&nbsp;куче становится не&nbsp;менее&nbsp;61. Победителем считается игрок, сделавший последний ход, т.&nbsp;е. первым получивший кучу из&nbsp;61 или более камней. В&nbsp;начальный момент в&nbsp;куче было <em>S</em>&nbsp;камней; <em>1&nbsp;≤&nbsp;S&nbsp;≤&nbsp;60</em>. Будем говорить, что игрок имеет выигрышную стратегию, если он&nbsp;может выиграть при любых ходах противника.</p>
<p>Укажите минимальное значение <em>S</em>, при котором Петя выигрывает первым ходом.</p></div>

<h3><details>
	<summary>Ответ</summary>
    31
</details></h3>
